```python
import pandas as pd

file_path = '/content/ZHVI.csv'
zillow_df = pd.read_csv(file_path)

zillow_df.head()
```

```python
zillow_df['Date'] = pd.to_datetime(zillow_df.iloc[:, 0])

# filtering for just 1/2020-5/2022
zillow_df_filtered = zillow_df[
    (zillow_df['Date'] >= '2020-01-01') & (zillow_df['Date'] <= '2022-05-31')
]
# dropping original date column to reset the index
zillow_df_filtered.reset_index(drop=True, inplace=True)

# new df 2020
display(zillow_df_filtered)
```

```python
# dropping the 'Unnamed: 0' column
zillow_df_filtered = zillow_df_filtered.drop(columns=zillow_df_filtered.columns[0])

zillow_df_filtered['Date'] = zillow_df_filtered['Date'].dt.strftime('%Y-%m')

# pivot the df
zillow_df_filtered = zillow_df_filtered.set_index('Date').T  # transpose it

# reset index to make the previous index (State names) a column
zillow_df_filtered.reset_index(inplace=True)

# rename the index column to 'State'
zillow_df_filtered = zillow_df_filtered.rename(columns={'index': 'State'})

# sorting the DataFrame by the 'State' column alphabetically
zillow_df_filtered = zillow_df_filtered.sort_values(by='State')

# fix indexing
zillow_df_filtered.reset_index(inplace=True)
zillow_df_filtered = zillow_df_filtered.drop(columns=["index"])
zillow_df_filtered.columns.name = None

display(zillow_df_filtered)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# check columns are not tuples and simplify column names
zillow_df_filtered.columns = [col[0] if isinstance(col, tuple) else col for col in zillow_df_filtered.columns]

# 'State' column already exists; no need to reset and rename the index
zillow_long = zillow_df_filtered.melt(id_vars=['State'], var_name='Date', value_name='Value')

# converting date to date time format
zillow_long['Date'] = pd.to_datetime(zillow_long['Date'], errors='coerce')

# dropping NaN values
zillow_long_cleaned = zillow_long.dropna(subset=['Date'])

# plotting the cleaned data
plt.figure(figsize=(16, 12))
sns.lineplot(data=zillow_long_cleaned, x='Date', y='Value', hue='State', palette='tab20', linewidth=2.0)

# add titles
plt.title('Zillow Home Value Trends for 2020-2022 by State', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Home Value ($)', fontsize=14)

plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')

# rotating x-axis ticks for better readability
plt.xticks(rotation=45)

# updated layout
plt.tight_layout()
plt.show()
```

```python
file_path = '/content/us-counties.csv'
covid_df = pd.read_csv(file_path)

display(covid_df)
```

```python
covid_df['date'] = pd.to_datetime(covid_df['date'])
# monthly data has to come from last day in a month, not aggregate every day
covid_df['month'] = covid_df['date'].dt.to_period('M')  # Extract month and year
last_day_cases = covid_df.sort_values(by='date').groupby(['state', 'month']).last().reset_index()
display(last_day_cases)
```

```python
# pivot the data to have months as columns and states as rows
monthly_cases_by_state = last_day_cases.pivot(index='state', columns='month', values='cases')

monthly_cases_by_state.columns.name = None
monthly_cases_by_state.reset_index(inplace=True)
monthly_cases_by_state = monthly_cases_by_state.rename(columns={'state': 'State'})
monthly_cases_by_state = monthly_cases_by_state.fillna(0)


# monthly_cases_by_state.columns = monthly_cases_by_state.columns.astype(str)
display(monthly_cases_by_state)
```

# need to move date column to front of df

```python
monthly_cases_by_state.columns = monthly_cases_by_state.columns.astype(str)

plt.figure(figsize=(15, 12))

for state in monthly_cases_by_state['State']:
    # row for the corresponding state
    state_data = monthly_cases_by_state[monthly_cases_by_state['State'] == state]

    # plotting the row's data (excluding the 'State' column)
    plt.plot(
        monthly_cases_by_state.columns[1:],  # Skip the 'State' column
        state_data.iloc[0, 1:],  # Get the first row as a Series, skipping 'State'
        label=state,
        alpha=0.6
    )

# titles and labels
plt.title("Monthly COVID-19 Cases by State (Last Day of Each Month)", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Cases", fontsize=12)

# rotating x-axis ticks for better readability
plt.xticks(rotation=45)

# legend
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title="States", fontsize=10)

plt.tight_layout()

plt.show()
```

### Covid visualization

```python
#creating new columns to calculate the avg covid cases in each state per year
def covid_yr_avg(data, year):
    covid_columns = [col for col in data.columns if str(year) in col and '_COVID' in col]
    data[covid_columns] = data[covid_columns].apply(pd.to_numeric, errors='coerce')
    data[f'{year}_Avg_COVID'] = data[covid_columns].mean(axis=1)

for year in [2020, 2021, 2022]:
    covid_yr_avg(viz_data, year)

viz_data.head()
```

```python
covid_columns_2020 = [col for col in viz_data.columns if '2020' in col and '_COVID' in col]
covid_columns_2021 = [col for col in viz_data.columns if '2021' in col and '_COVID' in col]
covid_columns_2022 = [col for col in viz_data.columns if '2022' in col and '_COVID' in col]

viz_data['Total_2020_COVID'] = viz_data[covid_columns_2020].sum(axis=1)
viz_data['Total_2021_COVID'] = viz_data[covid_columns_2021].sum(axis=1)

viz_data['Percent_Increase_2021_vs_2020'] = ((viz_data['Total_2021_COVID'] - viz_data['Total_2020_COVID']) /
                                         viz_data['Total_2020_COVID']) * 100

viz_data[['State', 'Total_2020_COVID', 'Total_2021_COVID',
      'Percent_Increase_2021_vs_2020']]
```

```python
#sort data by covid % increase
data_sorted = viz_data.sort_values('Percent_Increase_2021_vs_2020', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(
    x='Percent_Increase_2021_vs_2020',
    y='State',
    data=data_sorted,
    palette='coolwarm'
)

plt.title('Percent Increase in COVID-19 Cases (2021 vs 2020) by State', fontsize=16)
plt.xlabel('Percent Increase (%)', fontsize=12)
plt.ylabel('State', fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
