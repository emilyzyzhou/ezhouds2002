```python
import pandas as pd

file_path = '/content/ZHVI.csv'
zillow_df = pd.read_csv(file_path)

zillow_df.head()
```

```python
zillow_df['Date'] = pd.to_datetime(zillow_df.iloc[:, 0])

# filtering for just 1/2020-5/2022
zillow_df_filtered = zillow_df[
    (zillow_df['Date'] >= '2020-01-01') & (zillow_df['Date'] <= '2022-05-31')
]
# dropping original date column to reset the index
zillow_df_filtered.reset_index(drop=True, inplace=True)

# new df 2020
display(zillow_df_filtered)
```

```python
# dropping 'Unnamed: 0' column
zillow_df_filtered = zillow_df_filtered.drop(columns=zillow_df_filtered.columns[0])

zillow_df_filtered['Date'] = zillow_df_filtered['Date'].dt.strftime('%Y-%m')

# pivot
zillow_df_filtered = zillow_df_filtered.set_index('Date').T

display(zillow_df_filtered)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

zillow_df_filtered.columns = [col[0] if isinstance(col, tuple) else col for col in zillow_df_filtered.columns]

zillow_pivoted_reset = zillow_df_filtered.reset_index()
zillow_long = zillow_pivoted_reset.melt(id_vars=['index'], var_name='Date', value_name='Value')
zillow_long = zillow_long.rename(columns={'index': 'State'})

zillow_long['Date'] = pd.to_datetime(zillow_long['Date'], errors='coerce')

zillow_long_cleaned = zillow_long.dropna(subset=['Date'])

plt.figure(figsize=(16, 12))
sns.lineplot(data=zillow_long_cleaned, x='Date', y='Value', hue='State', palette='tab20', linewidth=2.0)

plt.title('Zillow Home Value Trends for 2020-2022 by State', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Home Value ($)', fontsize=14)
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
file_path = '/content/us-counties.csv'
covid_df = pd.read_csv(file_path)

display(covid_df)
```

```python
covid_df['date'] = pd.to_datetime(covid_df['date'])
```

```python
# monthly data has to come from last day in a month, not aggregate every day
covid_df['month'] = covid_df['date'].dt.to_period('M')  # Extract month and year
last_day_cases = covid_df.sort_values(by='date').groupby(['state', 'month']).last().reset_index()

# pivot the data to have months as columns and states as rows
monthly_cases_by_state = last_day_cases.pivot(index='state', columns='month', values='cases')

# monthly_cases_by_state.columns = monthly_cases_by_state.columns.astype(str)
display(monthly_cases_by_state)
```

```python
monthly_cases_by_state.columns = monthly_cases_by_state.columns.astype(str)

plt.figure(figsize=(15, 12))

for state in monthly_cases_by_state.index:
    plt.plot(monthly_cases_by_state.columns, monthly_cases_by_state.loc[state], label=state, alpha=0.6)

plt.title("Monthly COVID-19 Cases by State (Last Day of Each Month)", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Cases", fontsize=12)

plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title="States", fontsize=10)
plt.tight_layout()

plt.show()
```
